```python
import requests
from bs4 import BeautifulSoup

# Function to fetch and parse a sitemap
def fetch_and_parse_sitemap(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'xml')
        return [loc.text.strip() for loc in soup.find_all('loc')]
    else:
        print(f"Failed to fetch sitemap: Status code {response.status_code}")
        return []

# Defining the sitemap index URL
sitemap_index_url = 'https://techprimex.com/sitemap_index.xml'

# Fetching the sitemap index
response = requests.get(sitemap_index_url)
if response.status_code == 200:
    print("Sitemap index fetched successfully.")
    soup = BeautifulSoup(response.content, 'xml')

    # Extracting Sitemap URLs
    sitemap_urls = [sitemap_loc.text.strip() for sitemap_loc in soup.find_all('loc')]

    # Initializing a set to store all unique URLs
    all_urls = set()

    # Fetch and parse each listed sitemap
    for sitemap_url in sitemap_urls:
        print(f"Fetching sitemap: {sitemap_url}")
        urls = fetch_and_parse_sitemap(sitemap_url)
        all_urls.update(urls)  # Add URLs to the set

    # Save URLs to a file
    with open('urls.txt', 'w') as file:
        for url in all_urls:
            file.write(url + '\n')

    print("All URLs have been saved to 'scrappeddata.txt'.")

else:
    print(f"Failed to fetch sitemap index: Status code {response.status_code}")
```
